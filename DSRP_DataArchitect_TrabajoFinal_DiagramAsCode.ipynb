```python
# Importamos los recursos del Cloud Azure
# Fuente: https://diagrams.mingrammer.com/docs/nodes/azure

from diagrams import Diagram, Cluster
#Azure
from diagrams.azure.compute import AppServices
from diagrams.azure.analytics import DataFactories
from diagrams.azure.analytics import Databricks
from diagrams.azure.analytics import SynapseAnalytics
from diagrams.azure.ml import MachineLearningServiceWorkspaces
#AWS
from diagrams.aws.database import RDS
from diagrams.aws.business import BusinessApplications
from diagrams.aws.analytics import DataPipeline
from diagrams.aws.analytics import DataLakeResource
from diagrams.aws.analytics import Analytics
from diagrams.aws.ml import MachineLearning
#GCP
from diagrams.gcp.compute  import AppEngine
from diagrams.gcp.database import SQL
from diagrams.gcp.analytics import Dataproc
from diagrams.gcp.analytics import Bigquery
from diagrams.gcp.devtools import ToolsForVisualStudio
from diagrams.gcp.ml import AdvancedSolutionsLab


#Diagrama Architectura AZURE
with Diagram("\n Arquitectura de Datos con Azure") as diagAzure:
  with Cluster("Source"):
    app1 = AppServices("Structured Data - OLTP")
    app2 = AppServices("Semi-Structured Data")
    app3 = AppServices("Unstructured Data")

  ETL1 = DataFactories ("Batch")
  ETL2 = DataFactories ("Streaming")
  ETL3 = DataFactories ("Workflow Manager")

  with Cluster("LakeHouse"):
    bronceLayer = Databricks("Landing Zone")
    silverLayer = Databricks("Processing Zone")
    goldLayer = Databricks("Curated Data Zone")

    bronceLayer >> silverLayer >> goldLayer


  with Cluster ("Consumptions Zone"):
    dataMart = SynapseAnalytics("Dashboard")
    dataScience = MachineLearningServiceWorkspaces("Data Science")
    adHocQuery = MachineLearningServiceWorkspaces("Adhoc Query")
    consumers = [dataMart,dataScience, adHocQuery]

  app1 >> ETL1 >> bronceLayer
  app2 >> ETL2 >> bronceLayer
  app3 >> ETL3 >> bronceLayer

  goldLayer >> dataMart

  goldLayer >> dataScience

  bronceLayer >> adHocQuery
  silverLayer >> adHocQuery
  goldLayer >> adHocQuery
diagAzure

#Diagrama Architectura AWS
with Diagram("\n Arquitectura de Datos con AWS") as diagAWS:
  with Cluster("Source"):
    app1 = RDS("Structured Data - OLTP")
    app2 = BusinessApplications("Semi-Structured Data")
    app3 = BusinessApplications("Unstructured Data")

  ETL1 = DataPipeline ("Batch")
  ETL2 = DataPipeline ("Streaming")
  ETL3 = DataPipeline ("Workflow Manager")

  with Cluster("LakeHouse"):
    bronceLayer = DataLakeResource("Landing Zone")
    silverLayer = DataLakeResource("Processing Zone")
    goldLayer = DataLakeResource("Curated Data Zone")

    bronceLayer >> silverLayer >> goldLayer


  with Cluster ("Consumptions Zone"):
    dataMart = Analytics("Dashboard")
    dataScience = MachineLearning("Data Science")
    adHocQuery = Analytics("Adhoc Query")
    consumers = [dataMart,dataScience, adHocQuery]

  app1 >> ETL1 >> bronceLayer
  app2 >> ETL2 >> bronceLayer
  app3 >> ETL3 >> bronceLayer

  goldLayer >> dataMart

  goldLayer >> dataScience

  bronceLayer >> adHocQuery
  silverLayer >> adHocQuery
  goldLayer >> adHocQuery
diagAWS

#Diagrama Architectura GCP
with Diagram("\n Arquitectura de Datos con GCP") as diagGCP:
  with Cluster("Source"):
    app1 = SQL("Structured Data - OLTP")
    app2 = AppEngine("Semi-Structured Data")
    app3 = AppEngine("Unstructured Data")

  ETL1 = Dataproc ("Batch")
  ETL2 = Dataproc ("Streaming")
  ETL3 = Dataproc ("Workflow Manager")

  with Cluster("LakeHouse"):
    bronceLayer = Bigquery("Landing Zone")
    silverLayer = Bigquery("Processing Zone")
    goldLayer = Bigquery("Curated Data Zone")

    bronceLayer >> silverLayer >> goldLayer


  with Cluster ("Consumptions Zone"):
    dataMart = ToolsForVisualStudio("Dashboard")
    dataScience = AdvancedSolutionsLab("Data Science")
    adHocQuery = ToolsForVisualStudio("Adhoc Query")
    consumers = [dataMart,dataScience, adHocQuery]

  app1 >> ETL1 >> bronceLayer
  app2 >> ETL2 >> bronceLayer
  app3 >> ETL3 >> bronceLayer

  goldLayer >> dataMart

  goldLayer >> dataScience

  bronceLayer >> adHocQuery
  silverLayer >> adHocQuery
  goldLayer >> adHocQuery

diagGCP
diagAWS
```
